# Matplotlib

```
import matplotlib.pyplot as plt
```

A simple graph with pyplot:

```
plt.plot([1, 2, 3], [5, 125, 25])
```

Same graph, but in a more clear representation with title and labels:

```
x = [1, 2, 3]
y = [5, 125, 25]

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple graph")

plt.plot(x, y)
```

Now, we are adding second line to the graph, and legend as well:

```
x1 = [1, 2, 3]
y1 = [5, 125, 25]
x2 = [2, 8, 10]
y2 = [25, 255, 75]

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple graph")

plt.plot(x1, y1, label="blue line") #labels are added for legend
plt.plot(x2, y2, label="örange line")
plt.legend()
```
